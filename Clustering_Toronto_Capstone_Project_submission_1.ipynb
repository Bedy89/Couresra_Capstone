```python
Toronto_data= pd.read_csv('Toronto_Data.csv')
Toronto_data.head()
```

## 2. Let's clean the dataset 

```python
#Let's select the first three columns
Toronto_data= Toronto_data[['Column 1','Column 2','Column 3']]
Toronto_data
```

```python
Toronto_data.rename(columns={'Column 1':'PostalCode', 'Column 2':'Borough','Column 3':'Neighborhood'}, inplace=True)
Toronto_data.head()
```

```python
#Let's drop the first row
Toronto_data.dropna(axis=0,inplace=True)
Toronto_data.head()
```

```python
#Let's drop all the Postal Code with no assigned Borough and Neighborhood
Toronto_data=Toronto_data[Toronto_data.Borough!='Not assigned']
Toronto_data=Toronto_data.reset_index(drop=True)
Toronto_data
```

```python
# Let's check if there is any duplicared values
duplicateRowsDF = Toronto_data[Toronto_data.duplicated(['PostalCode'])]
print("Duplicate Rows based on a single column are:", duplicateRowsDF, sep='\n')
print('if we obtain an Empty DataFrame it means that there is no duplicate rows')
```
